```
# 3.1
seconds_in_hour = 60 * 60
print("Seconds in an hour:", seconds_in_hour)

# 3.2
seconds_per_hour = seconds_in_hour

# 3.3
seconds_in_day = seconds_per_hour * 24
print("Seconds in a day:", seconds_in_day)

# 3.4
seconds_per_day = seconds_in_day

# 3.5
float_division = seconds_per_day / seconds_per_hour
print("Seconds per day / seconds per hour (float division):", float_division)

# 3.6
integer_division = seconds_per_day // seconds_per_hour
print("Seconds per day // seconds per hour (integer division):", integer_division)
```
